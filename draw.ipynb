```python
import json

with open('results/agnews/llm_cbm/2025-11-14-01-30-16') as f:
    agnews_llm_cbm_alpha_1 = json.load(f)

with open('results/agnews/llm/2025-11-11-05-58-23') as f:
    agnews_llm = json.load(f)

with open('results/imdb/llm_cbm/2025-11-15-19-26-29') as f:
    imdb_llm_cbm_alpha_1 = json.load(f)

with open('results/imdb/llm/2025-11-15-20-15-58') as f:
    imdb_llm = json.load(f)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(np.array(agnews_llm_cbm_alpha_1) * 0.01, label='with rationale (alpha=1)')
plt.plot(np.array(agnews_llm) * 0.01, label='without rationale')    # x-axis is implicit: 0,1,2,3...
plt.xlabel("epoch")
plt.ylabel("accuracy")
plt.legend()
plt.title("Experimental Results")
txt="Dataset: AG News; Model: DistilBERT; Rationale: LLM; Selection: BADGE"
plt.figtext(0.5, -0.04, txt, wrap=True, horizontalalignment='center', fontsize=12)
plt.show()
"AG News LLM max accuracy:", max(agnews_llm), "\nAG News LLM_with_rationale_alpha_1 max accuracy:", max(agnews_llm_cbm_alpha_1) 
```
